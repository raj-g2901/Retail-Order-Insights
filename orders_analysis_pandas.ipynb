### Make the headings lower case and replace space with underscore


```python
df.columns=df.columns.str.lower()
# replace space with underscore
df.columns=df.columns.str.replace(' ','_') # Alt way - df.rename(columns={'Order Id':'order_id', 'City':'city'})
df.head(5)
```

### Derive new discount, sale price and profit column


```python
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']= df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

### convert order date from object data type to datetime


```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df['order_date']
```

### Dropping unneccesary columns- 'list_price','cost_price','discount_percent'


```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

### Loading Data into MySQL using `SQLAlchemy` engine

```python
#import pandas as pd
import sqlalchemy as sal

username = 'root'
password = '12345678'
host = 'localhost'
database = 'my_db'  # database name 

# Create a SQLAlchemy engine for MySQL
engine = sal.create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}/{database}')

# Create a connection
conn = engine.connect()
```

We use the `to_sql` function to load our cleaned and prepared dataset into a MySQL table.

```python

# load the data into MySQL
df.to_sql('df_orders', con=conn, index=False, if_exists='append') # df_orders is Table name
# here we are appending and not replacing because by deaful Pandas creates Table with highest value data types. which is not a good practice.
# So, we will create the database and table in MYSQL by ourselves of same names. And then append through pandas.
```
